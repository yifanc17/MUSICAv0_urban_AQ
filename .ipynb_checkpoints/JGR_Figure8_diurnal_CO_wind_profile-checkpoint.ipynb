```python
ctrl_co_day=xr.open_dataset(DATA_PATH+'CTRL_CO_day_240301.nc')
ctrl_co_night=xr.open_dataset(DATA_PATH+'CTRL_CO_night_240301.nc')
test_co_day=xr.open_dataset(DATA_PATH+'TEST_CO_day_240301.nc')
test_co_night=xr.open_dataset(DATA_PATH+'TEST_CO_night_240301.nc')
rural_co_day=xr.open_dataset(DATA_PATH+'RURAL_CO_day_240301.nc')
rural_co_night=xr.open_dataset(DATA_PATH+'RURAL_CO_night_240301.nc')
```

```python
ctrl_u_day=xr.open_dataset(DATA_PATH+'CTRL_U_day_240301.nc')
ctrl_u_night=xr.open_dataset(DATA_PATH+'CTRL_U_night_240301.nc')
test_u_day=xr.open_dataset(DATA_PATH+'TEST_U_day_240301.nc')
test_u_night=xr.open_dataset(DATA_PATH+'TEST_U_night_240301.nc')
rural_u_day=xr.open_dataset(DATA_PATH+'RURAL_U_day_240301.nc')
rural_u_night=xr.open_dataset(DATA_PATH+'RURAL_U_night_240301.nc')
```

```python
ctrl_v_day=xr.open_dataset(DATA_PATH+'CTRL_V_day_240301.nc')
ctrl_v_night=xr.open_dataset(DATA_PATH+'CTRL_V_night_240301.nc')
test_v_day=xr.open_dataset(DATA_PATH+'TEST_V_day_240301.nc')
test_v_night=xr.open_dataset(DATA_PATH+'TEST_V_night_240301.nc')
rural_v_day=xr.open_dataset(DATA_PATH+'RURAL_V_day_240301.nc')
rural_v_night=xr.open_dataset(DATA_PATH+'RURAL_V_night_240301.nc')
```

### near surface CO

```python
plt.rcParams['font.family'] = 'DejaVu Sans'
areaName='Thailand'
fig, axs = plt.subplots(1,4, figsize=(16, 6),
                        subplot_kw={'projection': ccrs.PlateCarree()})
plt.subplots_adjust(wspace=0.02)
plot1 = Plot_2D(rural_co_day*1e6,
                scrip_file=MUSICA_scrip_path,
                cmap='Spectral_r',
                cmin=0.1,
                cmax=2,
                lat_range=regions[areaName]['lat'],
                lon_range=regions[areaName]['lon'],
                title='RURAL DAY (LT 14:00)',
                colorbar=False,
                grid_line=False,
                font_family="DejaVu Sans",
                label_size=12,
                title_size=15,
                ax=axs[0])
add_se_streamlines(axs[0], plot1, rural_u_day, rural_v_day)

plot2 = Plot_2D(rural_co_night*1e6,
                scrip_file=MUSICA_scrip_path,
                cmap='Spectral_r',
                cmin=0.1,
                cmax=2,
                lat_range=regions[areaName]['lat'],
                lon_range=regions[areaName]['lon'],
                title='RURAL NIGHT (LT 02:00)',
                colorbar=False,
                grid_line=False,
                lonlat_line=False,
                font_family="DejaVu Sans",
                label_size=12,
                title_size=15,
                ax=axs[1])
add_se_streamlines(axs[1], plot2, rural_u_night, rural_v_night)

plot3 = Plot_2D(test_co_day*1e6,
                scrip_file=MUSICA_scrip_path,
                cmap='Spectral_r',
                cmin=0.1,
                cmax=2,
                lat_range=regions[areaName]['lat'],
                lon_range=regions[areaName]['lon'],
                title='URBAN DAY (LT 14:00)',
                colorbar=False,
                grid_line=False,
                lonlat_line=False,
                font_family="DejaVu Sans",
                label_size=12,
                title_size=15,
                ax=axs[2])
add_se_streamlines(axs[2], plot3, test_u_day, test_v_day)

plot4 = Plot_2D(test_co_night*1e6,
                scrip_file=MUSICA_scrip_path,
                cmap='Spectral_r',
                cmin=0.1,
                cmax=2,
                lat_range=regions[areaName]['lat'],
                lon_range=regions[areaName]['lon'],
                title='URBAN NIGHT (LT 02:00)',
                colorbar=False,
                grid_line=False,
                lonlat_line=False,
                font_family="DejaVu Sans",
                label_size=12,
                title_size=15,
                ax=axs[3])
add_se_streamlines(axs[3], plot4, test_u_night, test_v_night)

cb = fig.colorbar(plot1.im, ax=axs, orientation='vertical', 
                  shrink=0.8, pad=0.02, extend='both')
cb.ax.tick_params(labelsize=12)
cb.set_label(r'Near surfacec CO [$ppm$]', size=14)
plt.show()
```
